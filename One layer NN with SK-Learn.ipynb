```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split

plt.rcParams['figure.figsize'] = (15,5)
```

```python
n_feature  = 1
n_sample   = 1000

x=np.random.random([n_sample,n_feature])*2-1
y=np.sin(x[:,0])

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

plt.plot(x,y,'or')
plt.title('Plotting sinus data')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
clf = MLPRegressor(activation='tanh' ,max_iter=200, hidden_layer_sizes=(5,2),
              learning_rate='constant', learning_rate_init=0.001)
clf.fit(x_train, y_train)
clf.predict(x_test)
print('Accuracy on the training subset: {:.3f}'.format(clf.score(x_train, y_train)))
print('Accuracy on the test subset: {:.3f}'.format(clf.score(x_test, y_test)))
```
